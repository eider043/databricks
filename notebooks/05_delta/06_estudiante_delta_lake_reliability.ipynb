## 11. Comparar versión actual vs versión anterior

Esta consulta muestra la diferencia entre el estado inicial y el estado actual de la orden.

````python
%sql
WITH version_inicial AS (
  SELECT
    order_id,
    order_status AS order_status_version_0,
    delay_days AS delay_days_version_0
  FROM workspace.delta_lab.orders_reliability_lab VERSION AS OF 0
  WHERE order_id = (SELECT order_id FROM v_order_lab_candidate)
), version_actual AS (
  SELECT
    order_id,
    order_status AS order_status_actual,
    delay_days AS delay_days_actual
  FROM workspace.delta_lab.orders_reliability_lab
  WHERE order_id = (SELECT order_id FROM v_order_lab_candidate)
)
SELECT
  a.order_id,
  i.order_status_version_0,
  a.order_status_actual,
  i.delay_days_version_0,
  a.delay_days_actual
FROM version_actual a
JOIN version_inicial i
  ON a.order_id = i.order_id;
````

## Reto Nivel 2 — Time Travel y comparación

Comparación entre versión 0 (inicial) y versión actual de la tabla `orders_reliability_lab`:

````python
%sql
-- Reto Nivel 2: Comparación detallada entre versiones
WITH version_0 AS (
  SELECT 
    COUNT(*) as total_registros_v0,
    COUNT(DISTINCT order_status) as estados_distintos_v0,
    SUM(CASE WHEN order_status = 'canceled' THEN 1 ELSE 0 END) as cancelados_v0,
    MAX(delay_days) as max_delay_v0
  FROM workspace.delta_lab.orders_reliability_lab VERSION AS OF 0
),
version_actual AS (
  SELECT 
    COUNT(*) as total_registros_actual,
    COUNT(DISTINCT order_status) as estados_distintos_actual,
    SUM(CASE WHEN order_status = 'canceled' THEN 1 ELSE 0 END) as cancelados_actual,
    MAX(delay_days) as max_delay_actual
  FROM workspace.delta_lab.orders_reliability_lab
)
SELECT 
  v0.total_registros_v0,
  va.total_registros_actual,
  v0.cancelados_v0,
  va.cancelados_actual,
  (va.cancelados_actual - v0.cancelados_v0) as diferencia_cancelados,
  CASE 
    WHEN v0.total_registros_v0 = va.total_registros_actual 
         AND v0.cancelados_v0 = va.cancelados_actual
    THEN '✓ Versiones idénticas (RESTORE exitoso)'
    ELSE '✗ Hay diferencias entre versiones'
  END as estado_comparacion
FROM version_0 v0
CROSS JOIN version_actual va;
````

### Explicación de cambios:

**Resultado esperado**: Ambas versiones deben ser idénticas porque ejecutamos un RESTORE TO VERSION AS OF 0.

**¿Qué cambió durante el ejercicio?**
1. **Versión 0 → Versión 1**: Se ejecutó un UPDATE que cambió el `order_status` de una orden a 'canceled'
2. **Versión 1 → Versión 2**: Se ejecutó RESTORE que revirtió el cambio, volviendo al estado de la Versión 0

**Conclusión**: La comparación demuestra que RESTORE TABLE es una operación reversible que permite recuperar exactamente el estado anterior de los datos, sin pérdida de información.
